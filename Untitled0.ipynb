### Предобработка данных

Первый шаг — сгенерировать метки для обучающих данных, которые являются остаточным сроком службы (RUL), label1 и label2, как это было сделано в шаблоне Predictive Maintenance. Здесь мы будем использовать только «label1» для бинарной классификации, пытаясь ответить на вопрос: выйдет ли из строя конкретный двигатель в течение циклов w1?

```python
# Data Labeling - generate column RUL
rul = pd.DataFrame(train_df.groupby('id')['cycle'].max()).reset_index()
rul.columns = ['id', 'max']
train_df = train_df.merge(rul, on=['id'], how='left')
train_df['RUL'] = train_df['max'] - train_df['cycle']
train_df.drop('max', axis=1, inplace=True)
train_df.head()
```

```python
# generate label columns for training data
w1 = 30
w0 = 15
train_df['label1'] = np.where(train_df['RUL'] <= w1, 1, 0 )
train_df['label2'] = train_df['label1']
train_df.loc[train_df['RUL'] <= w0, 'label2'] = 2
train_df.head()
```

### Моделирование

Традиционные модели машинного обучения для прогнозирующего обслуживания основаны на разработке функций, которая заключается в ручном создании нужных функций с использованием знаний в предметной области и аналогичных методов. Это обычно затрудняет повторное использование этих моделей, поскольку разработка функций зависит от сценария проблемы и доступных данных, которые варьируются от предприятия к предприятию. Возможно, наиболее привлекательной частью применения глубокого обучения в области прогнозного обслуживания является тот факт, что эти сети могут автоматически извлекать нужные функции из данных, устраняя необходимость в ручном проектировании функций.

При использовании LSTM в области временных рядов одним из важных параметров, который необходимо выбрать, является длина последовательности, которая является окном для просмотра LSTM в обратном направлении. Это можно рассматривать как аналогию с выбором window_size = 5 циклов для вычисления характеристик подвижности в шаблоне прогнозирующего технического обслуживания, которые представляют собой среднее значение подвижности и стандартное отклонение подвижности для 21 значения датчика. Идея использования LSTMs заключается в том, чтобы позволить модели извлекать абстрактные характеристики из последовательности значений датчиков в окне, а не создавать их вручную. Предполагается, что если в этих значениях датчика есть закономерность в пределах окна, предшествующего отказу, то эта закономерность должна быть закодирована LSTM.

Одним из важнейших преимуществ LSTM является их способность запоминать долговременные последовательности (размеры окон), чего трудно достичь с помощью традиционной разработки функций. Например, вычисление скользящих средних по размеру окна в 50 циклов может привести к потере информации из-за сглаживания и абстрагирования значений за такой длительный период, вместо этого использование всех 50 значений в качестве входных данных может обеспечить лучшие результаты. В то время как разработка функций для окон больших размеров может не иметь смысла, LSTM могут использовать окна больших размеров и использовать всю информацию в окне в качестве входных данных. Ниже мы проиллюстрируем этот подход.

### LSTM

Слои Keras LSTM ожидают входных данных в виде массива numpy из 3 измерений (выборки, временные шаги, признаки), где выборки — это количество обучающих последовательностей, временные шаги — это окно ретроспективного анализа или длина последовательности, а признаки — это количество признаков каждой последовательности на каждом временном шаге.

```python
# функция для преобразования признаков в (выборки, временные шаги, признаки)
def gen_sequence(id_df, seq_length, seq_cols):
    '''Рассматриваются только последовательности, соответствующие длине окна, заполнение не используется.
       Это означает, что для тестирования нам нужно отбросить те, которые меньше длины окна.
       Альтернативой может быть заполнение последовательностей, чтобы мы могли использовать более короткие
    '''
    data_array = id_df[seq_cols].values
    num_elements = data_array.shape[0]
    for start, stop in zip(range(0, num_elements-seq_length), range(seq_length, num_elements)):
        yield data_array[start:stop, :]
```

```python
# собираем колонки с признаками
sensor_cols = ['s' + str(i) for i in range(1,22)]
sequence_cols = ['setting1', 'setting2', 'setting3', 'cycle_norm']
sequence_cols.extend(sensor_cols)
```

```python
# генератор для последовательностей
seq_gen = (list(gen_sequence(train_df[train_df['id']==id], sequence_length, sequence_cols))
           for id in train_df['id'].unique())
```

```python
# генерируем последовательности и конвертируем в массив numpy
seq_array = np.concatenate(list(seq_gen)).astype(np.float32)
seq_array.shape
```

```python
# функция для генерации лейблов
def gen_labels(id_df, seq_length, label):
    data_array = id_df[label].values
    num_elements = data_array.shape[0]
    return data_array[seq_length:num_elements, :]
```

```python
# генерируем лейблы
label_gen = [gen_labels(train_df[train_df['id']==id], sequence_length, ['label1'])
             for id in train_df['id'].unique()]
label_array = np.concatenate(label_gen).astype(np.float32)
label_array.shape
```

```python
# строим нейронную сеть, соединяя последовательно слои
nb_features = seq_array.shape[2]
nb_out = label_array.shape[1]

model_lstm = Sequential()

model_lstm.add(LSTM(
         input_shape=(sequence_length, nb_features),
         units=100,
         return_sequences=True))
model_lstm.add(Dropout(0.2))

model_lstm.add(LSTM(
          units=50,
          return_sequences=False))
model_lstm.add(Dropout(0.2))

model_lstm.add(Dense(units=nb_out, activation='sigmoid'))
model_lstm.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
print(model_lstm.summary())
```

```python
%%time
# fit the network
history_lstm = model_lstm.fit(seq_array, label_array, epochs=10, batch_size=200, validation_split=0.05, verbose=1,
          callbacks = [keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=0, patience=0, verbose=0, mode='auto')])
```

```python
import matplotlib.ticker as ticker

plt.plot(history_lstm.history['accuracy'])
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.plot(history_lstm.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()

plt.plot(history_lstm.history['loss'])
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.plot(history_lstm.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
scores_lstm_train = model_lstm.evaluate(seq_array, label_array, verbose=1, batch_size=200)
print('Accurracy: {}'.format(scores_lstm_train[1]))
```

```python
# прогнозируем и вычисляем матрицу неточности
y_pred_lstm_train = model_lstm.predict(seq_array, verbose=1, batch_size=200)
classes_x_lstm_train = np.round(y_pred_lstm_train)
y_true_train = label_array
print('Confusion matrix\n- x-axis is true labels.\n- y-axis is predicted labels')
cm_train = confusion_matrix(y_true_train, classes_x_lstm_train)
cm_train
```

```python
# вычисляем precision и recall
precision_lstm_train = precision_score(y_true_train, classes_x_lstm_train)
recall_lstm_train = recall_score(y_true_train, classes_x_lstm_train)
f1_test_lstm_train = 2 * (precision_lstm_train * recall_lstm_train) / (precision_lstm_train + recall_lstm_train)
print( 'precision =', precision_lstm_train, '\n', 'recall =', recall_lstm_train)
```
